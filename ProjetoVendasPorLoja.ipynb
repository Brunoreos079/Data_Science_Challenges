Desafio

Você faz parte da equipe de Analytics de uma grande marca de vestuário com mais de 25 lojas espalhadas em Shoppings de todo o Brasil.

Toda semana você precisa enviar para a diretoria um ranking atualizado com as 25 lojas contendo 3 informações:

Faturamento de cada Loja;
Quantidade de produtos vendidos de cada Loja;

Além disso, cada loja tem 1 gerente que precisa receber o resumo das informações da loja dele. Por isso, deve-se gerar um relatorio de cada loja contendo:

Ticket médio dos produto da sua Loja.

Esse relatório é sempre enviado como um resumo de todos os dados disponíveis no ano.

Aqui se inicia a parte de manipulação dos dados.

Para criar o dataframe com o faturamento total de cada loja eu:

Restringi meu dataframe dados a apenas duas colunas que me interessavam: ID Loja e Valor final.

Depois agrupei meus dados utilizando a função groupby e passando como parametro o ID Loja, depois somei estes dados e os ordenei de forma decrescente utilizando o "Valor Final" como parametro.

Assim temos um dataframe que demonstra qual a loja que teve maior faturamento no ano.

```python
dados_Fat_Total = dados[['ID Loja','Valor Final']].groupby('ID Loja').sum().sort_values('Valor Final',ascending= False)
display(dados_Fat_Total)
```

Para criar o dataframe com a quantidade total de produtos vendidos de cada loja eu:

Restringi meu dataframe dados a apenas duas colunas que me interessavam: ID Loja e Quantidade.

Depois agrupei meus dados utilizando a função groupby e passando como parametro o ID Loja, depois somei estes dados e os ordenei de forma decrescente utilizando a "Quantidade" como parametro.

Assim temos um dataframe que demonstra qual a loja que teve maior quantidade de produtos vendidos durante o ano.

```python
dados_Total_Prod = dados[['ID Loja','Quantidade']].groupby('ID Loja').sum().sort_values('Quantidade',ascending= False)
display(dados_Total_Prod)
```

Para criar o dataframe com o Ticket médio de cada loja eu:

Criei um dataframe apartir da divisão do meu faturamento total pela minha quantidade total vendida, ambas as colunas obtidas dos dataframes gerados anteriormente.

Depois ordenei meu dados de forma decrescente utilizando o "Ticket Medio" como parametro.

Assim temos um dataframe que demonstra qual a loja que teve o maior Ticket Medio no ano.

```python
dados_Ticket_Medio = (dados_Fat_Total['Valor Final'] / dados_Total_Prod['Quantidade']).to_frame('Ticket Medio')
dados_Ticket_Medio = dados_Ticket_Medio.sort_values('Ticket Medio',ascending = False)
display(dados_Ticket_Medio)
```

Agora para realizar a parte do desafio de gerar os relatórios semanais.

Para tornar mais amplo, decidi criar uma função que realizasse a geração de relatorio de qualquer periodo.

Comecei importando a biblioteca datetime


```python
from datetime import datetime
```

Aqui foi criado a função geral_Rel que gera os meus relatório, ela recebe como parametro um IDLoja, uma data de inicio e uma data de fim.

Ela recebe estas datas em string e as converte em datetime utilizando a função strptime, e após isso é criado um dataframe chamado relatorio que é prenchido através da localização (função loc) das linhas do meu dataframe que correspondem a Loja desejada e no periodo de tempo desejado

```python
def gerar_Rel(IDLoja1, Inicio, Fim):

              Inicio = datetime.strptime(Inicio,'%Y/%m/%d')
              Fim = datetime.strptime(Fim,'%Y/%m/%d')
              relatorio = dados.loc[(dados['ID Loja'] == IDLoja1)&(dados.Data >= Inicio) & (dados.Data <= Fim)]
              return relatorio
```

```python
rel_NS_Semana1_Fev = gerar_Rel('Norte Shopping','2019/2/1','2019/2/7')
```

Aqui foi feito um teste em que foi criado um dataframe representando o relatorio da Loja Norte Shopping na primeira semana de fevereiro de 2019.

```python
rel_NS_Semana1_Fev
```
